```python
# Shapiro-Wilk Test
from numpy.random import seed
from numpy.random import randn
from scipy.stats import shapiro
print("//// Shapiro-Wilk Test - WT Clones ////")
# generate univariate observations
data = [wt_b6_doubling_time, wt_b7_doubling_time , wt_c10_doubling_time, wt_f4_doubling_time, wt_f9_doubling_time , wt_g5_doubling_time]
# normality test
stat, p = shapiro(data)
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
	print('Sample looks Gaussian (fail to reject H0)')
else:
	print('Sample does not look Gaussian (reject H0)')
```

```python
print("//// Shapiro-Wilk Test - DC Clones ////")

data = [dc_f5_doubling_time, dc_f6_doubling_time, dc_g4_doubling_time, dc_d5_doubling_time, dc_e2_doubling_time]
stat, p = shapiro(data)
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
    print('Sample looks Gaussian (fail to reject H0)')
else:
    print('Sample does not look Gaussian (reject H0)')
```

As you can see, both samples seem to be normally distributed. Now, we can perform an two-sample t-test for the null hypothesis that there is not a significant difference in doubling times between both samples. This is a two-sided test and assumes that the two samples are indepeandant and normally distributed. 

```python
from __future__ import print_function

import numpy as np
from scipy.stats import ttest_ind, ttest_ind_from_stats
from scipy.special import stdtr
a = [dc_f5_doubling_time, dc_f6_doubling_time, dc_g4_doubling_time, dc_d5_doubling_time, dc_e2_doubling_time]
b = [wt_b6_doubling_time, wt_b7_doubling_time , wt_c10_doubling_time, wt_f4_doubling_time, wt_f9_doubling_time , wt_g5_doubling_time]
t, p = ttest_ind(a, b, equal_var=False)
```

```python
print("/// Two Sample T Test /// \n")
import numpy as np
from scipy.stats import ttest_ind, ttest_ind_from_stats
from scipy.special import stdtr

a = np.array(a)
b = np.array(b)

# Compute the descriptive statistics of a and b.
abar = a.mean()
avar = a.var(ddof=1)
na = a.size
adof = na - 1

bbar = b.mean()
bvar = b.var(ddof=1)
nb = b.size
bdof = nb - 1

t2, p2 = ttest_ind_from_stats(abar, np.sqrt(avar), na,
                              bbar, np.sqrt(bvar), nb,
                              equal_var=False)
print("ttest_ind_from_stats: t = %g  p = %g \n" % (t2, p2))

print("Using the formulas directly:")
tf = (abar - bbar) / np.sqrt(avar/na + bvar/nb)
dof = (avar/na + bvar/nb)**2 / (avar**2/(na**2*adof) + bvar**2/(nb**2*bdof))
pf = 2*stdtr(dof, -np.abs(tf))

print("formula:              t = %g  p = %g" % (tf, pf))
```

Because the p value is much higher that alpha (.05), we fail the reject the hypothesis that these two samples are significantly different in their doubling times. Because this is a limited data set, the power of the Shapiro-Wilk test to determine normality is significantly limited. We can also perform a Mann-Whitney U test for the same null hypothesis. The Mann-Whitney U test is similar to a two sample t test execpt for the fact that it does not assume normality for the two data sets. 

```python
import numpy as np
import scipy.stats as stats
group1 = [dc_f5_doubling_time, dc_f6_doubling_time, dc_g4_doubling_time, dc_d5_doubling_time, dc_e2_doubling_time]

group2 = [wt_b6_doubling_time, wt_b7_doubling_time , wt_c10_doubling_time, wt_f4_doubling_time, wt_f9_doubling_time , wt_g5_doubling_time]
u_statistic, pVal = stats.mannwhitneyu(group1, group2)
print ('Mann Whitney U Test // U-value: ' + str(u_statistic) + " P-Value: " + str(pVal))
```

The results of the Mann-Whitnery U-Test valdiate the two-sample t-test. We have failed to reject the null.
